```python
#Task 01: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
  #TO DO
  result = np.array([0]*len(pokemon_1))
  n=0
  while n < len(pokemon_1):
    if pokemon_1[n]==None and pokemon_2[(len(pokemon_1)-1-n)]!= None:
      result[n]= pokemon_2[len(pokemon_1)-1-n]
    elif pokemon_2[(len(pokemon_1)-1-n)]==None and pokemon_1[n]!= None:
      result[n]=pokemon_1[n]
    elif pokemon_1[n]!= None and pokemon_2[(len(pokemon_1)-1-n)]!= None:
      result[n]= pokemon_1[n]+pokemon_2[len(pokemon_1)-1-n]
    else:
      result[n] = None
    n+=1
  return result


print("///  Task 01: Merge Lineup  ///")
pokemon_1 = np.array([12, 3, 25, 1, None])
pokemon_2 = np.array([5, -9, 3, None, None] )
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 1: {returned_value}') # This should print [12, 3, 28, -8, 5]
unittest.output_test(returned_value, np.array([12, 3, 28, -8, 5]))

pokemon_1 = np.array([4, 5, -1, None, None])
pokemon_2 = np.array([2, 27, 7, 12, None])
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 1: {returned_value}') # This should print [4,17,6,27,2]
unittest.output_test(returned_value, np.array([4,17,6,27,2]))
```

```python
# Task 03: DUBER Fare Splitting
def findGroups(money, fare):
  # TO DO
  ungroup =""
  count = 1

  for i in range(len(money)):
    sum = money[i]
    if sum==fare:
      print(f"Group {count} : {sum}")
      money[i] = 0
      count += 1
    else:
      for j in range(1+i,len(money)):
        sum= money[i] + money[j]

        if sum == fare:
          print(f"Group {count} : {money[i]}, {money[j]}")
          money[i],money[j]=0,0
          count += 1
          break

  for i in range(len(money)):

    if money[i] !=0:
      ungroup += f"{money[i]} "

  if len(ungroup)>0:
    print(f"ungrouped : {ungroup}")


print("///  Task 03: DUBER Fare Splitting  ///")
money = np.array( [120, 100, 150, 50, 30])
fare = 150
print(f'Task 3:')
findGroups(money, fare) # This should print

# Group 1 : 120, 30
# Group 2 : 100, 50
# Group 3 : 150


money = np.array( [60, 150, 60, 30, 120, 30])
fare = 180
print(f'Task 3:')
findGroups(money, fare) # This should print

# Group 1 : 60, 120
# Group 2 : 30, 150
# Ungrouped : 30 60
```

```python
# Task 04: Get Those Hobbies
def analyzeHobbies(* participants):
  total = 0

  for i in range(len(participants)):
    total += len(participants[i])
  new = np.array([None]*total)

  ind = 0
  for i in range(len(participants)):
    for j in range(participants[i].size):
      new[ind] = participants[i][j]
      ind +=1

  ind = 0
  unique= np.array([None]*total)

  for i in range(len(new)):
    if new[i] not in unique:
      unique[ind] = new[i]
      ind += 1

  new_unique = np.array([None]*ind)

  for i in range(len(new_unique)):
    new_unique[i] = unique[i]

  print(f"Unique Activities in the Town:\n{new_unique}\n")
  print("Statistics:")

  for i in range(len(new_unique)):
    count = 0
    for j in range(len(new)):
      if new_unique[i] == new[j]:
        count+=1
    print(f"{count}participant(s) like(s) {new_unique[i]}")


print("///  Task 04: Get Those Hobbies  ///")
participant_1 = np.array( ["Hiking", "Reading", "Photography", "Cooking"])
participant_2 = np.array( ["Reading", "Hiking", "Painting"])
participant_3 = np.array( ["Hiking", "Cooking", "Photography"])
print(f'Task 4:')
analyzeHobbies(participant_1, participant_2, participant_3) #This should print

#Unique Activities in the Town:
#['Photography', 'Painting', 'Cooking', 'Reading', 'Hiking']

#Statistics:
#2 participant(s) like(s) Photography.
#1 participant(s) like(s) Painting.
#2 participant(s) like(s) Cooking.
#2 participant(s) like(s) Reading.
#3 participant(s) like(s) Hiking.



participant_1 = np.array( ["Gardening", "Traveling"])
participant_2 = np.array( ["Singing", "Gardening", "Painting"])
print(f'Task 4:')
analyzeHobbies(participant_1, participant_2) #This should print

#Unique Activities in the Town:
#[Gardening, Traveling, Singing, Painting]

#Statistics:
#2 participant(s) like(s) Gardening.
#1 participant(s) like(s) Traveling.
#1 participant(s) like(s) Singing.
#1 participant(s) like(s) Painting.
```

```python
import math
import numpy as np

def calculate_mean(arr):
    sum = 0

    for i in range(len(arr)):
        sum += arr[i]
    mean_value = sum / len(arr)

    return mean_value

def calculate_standard_deviation(arr):
    deviation_sum = 0
    mean_value = calculate_mean(arr)

    for i in range(len(arr)):
        deviation_sum += (arr[i] - mean_value) ** 2

    standard_deviation_value = math.sqrt((1 / (len(arr) - 1)) * deviation_sum)

    return standard_deviation_value

def total_calculation(arr):
    mean_value = calculate_mean(arr)
    standard_deviation_value = calculate_standard_deviation(arr)

    val_1 = mean_value + (1.5 * standard_deviation_value)
    val_2 = mean_value - (1.5 * standard_deviation_value)

    c = 0
    for i in range(len(arr)):
        if float(arr[i]) < val_2 or float(arr[i]) > val_1:
            c += 1
    new_arr = np.zeros(c, dtype=int)
    d = 0
    for i in range(len(arr)):
        if float(arr[i]) < val_2 or float(arr[i]) > val_1:
            new_arr[d] = arr[i]
            d += 1

    if len(new_arr) == 0:
      return f"The mean of the numbers is: {mean_value}\n\nThe standard deviation is:{standard_deviation_value:.2f}\n\nNone"


    return f"The mean of the numbers is: {mean_value}\n\nThe standard deviation is:{standard_deviation_value:.2f}\n\nNew array:{new_arr}"

arr = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
print(total_calculation(arr))
```
