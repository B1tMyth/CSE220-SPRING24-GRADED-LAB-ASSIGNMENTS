```python
#Task 01: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
  # TO DO
  result = np.array([0]*len(pokemon_1))
  j=len(pokemon_1)-1
  for i in range(0,len(pokemon_1)):
    if(pokemon_1[i]==None):
      pokemon_1[i]=0
    if(pokemon_2[j-i]==None):
      pokemon_2[j-i]=0
    result[i]=pokemon_1[i]+pokemon_2[j-i]
  return result

print("///  Task 01: Merge Lineup  ///")
pokemon_1 = np.array([12, 3, 25, 1, None])
pokemon_2 = np.array([5, -9, 3, None, None] )
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 1: {returned_value}') # This should print [12, 3, 28, -8, 5]
unittest.output_test(returned_value, np.array([12, 3, 28, -8, 5]))

pokemon_1 = np.array([4, 5, -1, None, None])
pokemon_2 = np.array([2, 27, 7, 12, None])
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 1: {returned_value}') # This should print [4,17,6,27,2]
unittest.output_test(returned_value, np.array([4,17,6,27,2]))
```

```python
# Task 02: Discard Cards
def discardCards(cards, t):
  # TO DO
  count=-1
  for i in range(0,len(cards)):
    if(cards[i]==t):
      count+=1

      if(count//2==0):
         left_shift(cards,i)
  return cards

def left_shift(arr,index):
  for i in range(index,len(arr)-1,1):
    arr[i]=arr[i+1]
  arr[len(arr)-1]=0
  return arr


print("///  Task 02: Discard Cards  ///")
cards = np.array([1,3,7,2,5,2,2,2,0])
returned_value = discardCards(cards, 2)
print(f'Task 2: {returned_value}') # This should print [1,3,7,5,2,2,0,0,0]
unittest.output_test(returned_value, np.array([1,3,7,5,2,2,0,0,0]))

cards = np.array([5,5,5,0,0])
returned_value = discardCards(cards, 5)
print(f'Task 2: {returned_value}') # This should print [5,0,0,0,0]
unittest.output_test(returned_value, np.array([5,0,0,0,0]))
```

```python
# Task 03: DUBER Fare Splitting
def findGroups(money, fare):
  # TO DO
  # Print outputs inside the method
  count=0
  temp=0
  co=0
  select=np.array([0]*(len(money)))
  Ung=np.array([0]*len(money))
  for i in range(0,len(select),1):
    count+=1
    if(select[i]==0):
      temp=fare-money[i]
      if(money[i]==fare):
        select[i]=1
        co+=1
        print(f'Group {count} : {money[i]}')

      else:
        for j in range(0,len(select)):
            if(money[j]==temp and select[j]==0):
              select[i]=1
              select[j]=1
              co+=2
              print(f'Group {count} : {money[i]},{money[j]}')
              break
  Ung=np.array([0]*(len(money)-co))
  if(len(Ung)>0):
    s=0
    for k in range(0,len(select)):
      if(select[k]==0):
          Ung[s]=money[k]
          s+=1
    print(f'Ungrouped :', end='')
    for i in Ung:
      print(i, end = " ")

print("///  Task 03: DUBER Fare Splitting  ///")
money = np.array( [120, 100, 150, 50, 30])
fare = 150
print(f'Task 3:')
findGroups(money, fare) # This should print

# Group 1 : 120, 30
# Group 2 : 100, 50
# Group 3 : 150


money = np.array( [60, 150, 60, 30, 120, 30])
fare = 180
print(f'Task 3:')
findGroups(money, fare) # This should print

# Group 1 : 60, 120
# Group 2 : 30, 150
# Ungrouped : 30 60
```

```python
import numpy as np
import math
def calculate_mean(arr):
  sum=0
  for i in arr:
    sum+=i
  return (sum/(len(arr)))
def standard_deviation(arr):
  sum1=0
  sd=0
  temp=calculate_mean(arr)
  for i in arr:
    sum1+=(i-temp)**2
  sd=math.sqrt(sum1/(len(arr)-1))
  return sd
def new_array(arr,standard_value):
  Up=calculate_mean(arr)+standard_value*standard_deviation(arr)
  Lo=calculate_mean(arr)-standard_value*standard_deviation(arr)
  count=0
  for i in arr:
    if(i>Up):
      count+=1
    if(i<Lo):
      count+=1
  new=np.array([0]*count)
  count=0
  for j in arr:
    if(j>Up):
      new[count]=j
      count+=1
    if(j<Lo):
      new[count]=j
      count+=1
  return new
arr1=np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
print(calculate_mean(arr1))
print(standard_deviation(arr1))
print(new_array(arr1,1.5))
```
