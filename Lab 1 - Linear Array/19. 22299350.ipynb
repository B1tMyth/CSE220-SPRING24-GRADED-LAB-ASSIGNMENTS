```python
#Task 01: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
  # TO DO
  result = [None]*len(pokemon_1)
  for i in range(len(pokemon_1)):
    if pokemon_1[i]==None:
      pokemon_1[i]=0
  for j in range(len(pokemon_1)):
    if pokemon_2[j]==None:
      pokemon_2[j]=0
  for i in range(len(pokemon_1)):
    val=pokemon_1[i]+pokemon_2[len(pokemon_2)-1-i]
    result[i]=val
  return np.array(result)

print("///  Task 01: Merge Lineup  ///")
pokemon_1 = np.array([12, 3, 25, 1, None])
pokemon_2 = np.array([5, -9, 3, None, None] )
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 1: {returned_value}') # This should print [12, 3, 28, -8, 5]
unittest.output_test(returned_value, np.array([12, 3, 28, -8, 5]))

pokemon_1 = np.array([4, 5, -1, None, None])
pokemon_2 = np.array([2, 27, 7, 12, None])
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 1: {returned_value}') # This should print [4,17,6,27,2]
unittest.output_test(returned_value, np.array([4,17,6,27,2]))
```

```python
import numpy as np
def mean (arr):
  total=0
  for v in range (len(arr)):
    total+=arr[v]
  mean = total/len(arr)
  return mean

def standard_deviation(arr):
  total=0
  count=0
  for v in range (len(arr)):
    count+=1
    total+=((arr[v]-mean(arr))*(arr[v]-mean(arr)))
  standardDeviation = (total/(count-1))**0.5
  return standardDeviation

def notun_array(arr):
  val1 = mean(arr) + (1.5* standard_deviation(arr))
  val2= mean (arr) - (1.5* standard_deviation(arr))
  length=0
  for v in range (len(arr)):
    if arr[v] > val1 or arr[v]< val2:
      length+=1
  notun_array= np.array([0]*length)
  t=0
  for v in range(len(arr)):
    if arr[v]> val1 or arr[v] < val2:
      notun_array[t]= arr[v]
      t+=1
  return f"The mean of the number is :{mean(arr)}\nThe standard deviation: {round(standard_deviation(arr),2)}\nNew array : {notun_array}"

arr =np.array ([10, 8, 13, 9, 14, 25, -5, 20, 7, 7,4])
print(notun_array(arr))
```
