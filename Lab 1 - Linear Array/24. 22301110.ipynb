```python
# part 1
# (Playing with Array and Probability Theory)
import numpy as np
def mean_(arr):
  s = 0
  for i in range(len(arr)):
    s = s + arr[i]
  s = s/len(arr)
  print(s)
  return s

def standard_deviation_(arr):
  mean = mean_(arr)
  s = 0
  for i in range(len(arr)):
    s = s + (arr[i]-mean)**2
  s = s / len(arr)
  s = s**0.5
  print(s)
  return s

def new_arr_(arr):
  mean = mean_(arr)
  sta_de = standard_deviation_(arr)
  element_counter = 0
  for i in range(len(arr)):
    if arr[i] >= mean + (1.5*sta_de) or arr[i] <= mean - (1.5*sta_de):
      element_counter = element_counter + 1

  new = np.array([0]*element_counter)
  c = 0
  for i in range(len(arr)):
    if arr[i] >= mean + (1.5*sta_de) or arr[i] <= mean - (1.5*sta_de):
      new[c] = arr[i]
      c += 1
  return new

a = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
new_arr = new_arr_(a)
print(new_arr)
```

```python
# Task 02: Discard Cards

def discardCards(cards,number):
  c = 0
  x = 0
  for i in range(len(cards)):
    if cards[i] == number:
      c += 1
  new = [0]*c
  c = 0
  for i in range(len(cards)):
    if cards[i] == number:
      new[c] = i
      c += 1
  for i in range(len(new)):
    for j in range(new[i]-x,len(cards)-1):
      cards[j] = cards[j+1]
      cards[j+1] = 0
    x += 1
  return cards




print("///  Task 02: Discard Cards  ///")
cards = np.array([1,3,7,2,5,2,2,2,0])
returned_value = discardCards(cards, 2)
print(f'Task 2: {returned_value}') # This should print [1,3,7,5,2,2,0,0,0]
unittest.output_test(returned_value, np.array([1,3,7,5,2,2,0,0,0]))

cards = np.array([5,5,5,0,0])
returned_value = discardCards(cards, 5)
print(f'Task 2: {returned_value}') # This should print [5,0,0,0,0]
unittest.output_test(returned_value, np.array([5,0,0,0,0]))
```
