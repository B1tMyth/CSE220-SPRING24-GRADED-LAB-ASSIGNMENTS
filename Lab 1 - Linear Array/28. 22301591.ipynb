```python
#Task 01: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
  result = np.array([0]*len(pokemon_1))
  #to do
  for i in range(len(pokemon_1)):
    if pokemon_1[i] is not None:
      result[i] += pokemon_1[i]

    if pokemon_2[-(i+1)] is not None:
      result[i] += pokemon_2[-(i+1)]
  return result
print("///  Task 01: Merge Lineup  ///")
pokemon_1 = np.array([12, 3, 25, 1, None])
pokemon_2 = np.array([5, -9, 3, None, None] )
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 1: {returned_value}') # This should print [12, 3, 28, -8, 5]
unittest.output_test(returned_value, np.array([12, 3, 28, -8, 5]))
pokemon_1 = np.array([4, 5, -1, None, None])
pokemon_2 = np.array([2, 27, 7, 12, None])
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 1: {returned_value}') # This should print [4,17,6,27,2]
unittest.output_test(returned_value, np.array([4,17,6,27,2]))
```

```python
# Task 02: Discard Cards

def discardCards(cards, t):
  # TO DO
  i = 0
  while i < len(cards):
    if cards[i] == t:
      cards = np.delete(cards, i)
    else:
      i += t - 1
    i += 1
  cards = np.append(cards, [0] * (len(cards) - len(np.trim_zeros(cards))))

  return cards


print("///  Task 02: Discard Cards  ///")
cards = np.array([1,3,7,2,5,2,2,2,0])
returned_value = discardCards(cards, 2)
print(f'Task 2: {returned_value}') # This should print [1,3,7,5,2,2,0,0,0]
unittest.output_test(returned_value, np.array([1,3,7,5,2,2,0,0,0]))

cards = np.array([5,5,5,0,0])
returned_value = discardCards(cards, 5)
print(f'Task 2: {returned_value}') # This should print [5,0,0,0,0]
unittest.output_test(returned_value, np.array([5,0,0,0,0]))
```

```python
# Task 03: DUBER Fare Splitting
def findGroups(money, fare):
  # TO DO
  # Print outputs inside the method
  ungrouped = np.array([])

  for i in range(len(money)):
    if money[i] != 0:
      for j in range(i + 1, len(money)):
        if money[j] != 0 and money[i] + money[j] == fare:
          print(f'Group: {money[i]}, {money[j]}')
          money[i] = 0
          money[j] = 0
          break

  for k in money:
    if k != 0:
      ungrouped = np.append(ungrouped, k)

  if len(ungrouped) > 0:
    print(f'Ungrouped: {ungrouped}')

print("///  Task 03: DUBER Fare Splitting  ///")
money = np.array( [120, 100, 150, 50, 30])
fare = 150
print(f'Task 3:')
findGroups(money, fare) # This should print

# Group 1 : 120, 30
# Group 2 : 100, 50
# Group 3 : 150


money = np.array( [60, 150, 60, 30, 120, 30])
fare = 180
print(f'Task 3:')
findGroups(money, fare) # This should print

# Group 1 : 60, 120
# Group 2 : 30, 150
# Ungrouped : 30 60
```

```python
from array import array
import math

def calculate_mean(arr):
    total = 0
    for num in arr:
        total += num
    return total / len(arr)

def calculate_standard_deviation(arr, mean):
    squared_diff_sum = 0
    for num in arr:
        squared_diff_sum += (num - mean) ** 2
    variance = squared_diff_sum / len(arr)
    return math.sqrt(variance)

def filter_values_1_5_std_away(arr, mean, std_dev):
    hold = 1.5 * std_dev
    filtered_values = array('h', [x for x in arr if abs(x - mean) >= hold])
    return filtered_values

# Driver code
input_array = array('h', [10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])

mean_value = calculate_mean(input_array)
std_dev_value = calculate_standard_deviation(input_array, mean_value)
new_array = filter_values_1_5_std_away(input_array, mean_value, std_dev_value)

print(f"The mean of the numbers is: {mean_value}")
print(f"The standard deviation is: {std_dev_value:.2f}")
print(f"New array: {new_array}")
```
