```python
#Task 01: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
  # TO DO
  pok1 = pokemon_1
  pok2 = pokemon_2
  result = np.array([0]*len(pokemon_1))

  for i in range(len(pok1)):
    if pok1[i]==None:
      result[i]+=0
    else:
      result[i]+=pok1[i]

  count= 0
  for j in range((len(pok2)-1), -1 ,-1):
    if pok2[j]==None:
      result[count]+= 0

    else:
      result[count]+=pok2[j]
    #print(result)

    count+=1

  return result



print("///  Task 01: Merge Lineup  ///")
pokemon_1 = np.array([12, 3, 25, 1, None])
pokemon_2 = np.array([5, -9, 3, None, None] )
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 1: {returned_value}') # This should print [12, 3, 28, -8, 5]
unittest.output_test(returned_value, np.array([12, 3, 28, -8, 5]))

pokemon_1 = np.array([4, 5, -1, None, None])
pokemon_2 = np.array([2, 27, 7, 12, None])
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 1: {returned_value}') # This should print [4,17,6,27,2]
unittest.output_test(returned_value, np.array([4,17,6,27,2]))
```

```python
# Task 02: Discard Cards
def discardCards(cards, t):
  # TO DO
  m = 0
  for i in cards:
    if i == t:
      m+= 1
  x = np.array([0]*m)

  count = 0
  for i in range(len(cards)):
    if cards[i] == t:
      x[count] = i
      count+= 1

      flag= False

  if len(x)%2 == 0:

    for j in range(0,len(x)-1,2):

      if flag == False:
        y = x[j]
        for k in range(y,(len(cards)-1)):
          cards[k],cards[k+1] = cards[k+1],0
        flag = True

      else:
        y = x[j]
        for k in range(y,len(cards)-1):
          cards[k],cards[k+1] = cards[k+1],0


  else:
    for j in range(0,len(x),2):
      if flag == False:
        y = x[j]
        for k in range(y,len(cards)-1):
          cards[k],cards[k+1] = cards[k+1],0
        flag = True

      else:
        y = x[j]-1
        for k in range(y,len(cards)-1):
          cards[k],cards[k+1] = cards[k+1],0

  return cards

print("///  Task 02: Discard Cards  ///")
cards = np.array([1,3,7,2,5,2,2,2,0])
returned_value = discardCards(cards, 2)
print(f'Task 2: {returned_value}') # This should print [1,3,7,5,2,2,0,0,0]
unittest.output_test(returned_value, np.array([1,3,7,5,2,2,0,0,0]))
cards = np.array([5,5,5,0,0])
returned_value = discardCards(cards, 5)
print(f'Task 2: {returned_value}') # This should print [5,0,0,0,0]
unittest.output_test(returned_value, np.array([5,0,0,0,0]))
```

```python
# Task 03: DUBER Fare Splitting
def findGroups(money, fare):
  # TO DO
  grp = ""
  count = 1
  for i in range(len(money)):
    sum = money[i]
    if sum == fare:
      print(f"Group: {count}: {sum}")
      money[i] = 0
      count+=1


    else:
      for j in range(i+1,len(money)):

          sum= money[i]+money[j]
          if sum == fare:
            print(f"Group: {count}: {money[i]},  {money[j]}")
            money[i] = 0
            money[j] = 0
            count+=1
            break


  for k in range(len(money)):
    if money[k]!=0:
      grp+= f"{money[k]} "

  if len(grp)>0:
    print(f"Ungrouped: { grp } ")

  # Print outputs inside the method

print("///  Task 03: DUBER Fare Splitting  ///")
money = np.array( [120, 100, 150, 50, 30])
fare = 150
print(f'Task 3:')
findGroups(money, fare) # This should print

# Group 1 : 120, 30
# Group 2 : 100, 50
# Group 3 : 150


money = np.array( [60, 150, 60, 30, 120, 30])
fare = 180
print(f'Task 3:')
findGroups(money, fare) # This should print

# Group 1 : 60, 120
# Group 2 : 30, 150
# Ungrouped : 30 60
```

```python
def mean(array):
  sum = 0
  count = 0
  while count!= len(array):
    sum+= array[count]
    count+= 1
  mean = sum/count
  return mean

def standard_deviation(array,mean):
  new_array = np.zeros(2,int)
  count = 0
  sum = 0
  while count!= len(array):
    sum+= (array[count]-mean)**2
    count+=1
  sum= (sum/(len(array)-1))**(1/2)
  j = 0

  for i in range(len(array)):
    if (array[i]-mean)>= 1.5*sum:
      new_array[j]=array[i]
      j+=1
  return f"The standard deviation is: {sum.round(2)}\nNew array: {new_array}"

x=np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
mean = mean(x)
print(mean)
print(standard_deviation(x,mean))
```
