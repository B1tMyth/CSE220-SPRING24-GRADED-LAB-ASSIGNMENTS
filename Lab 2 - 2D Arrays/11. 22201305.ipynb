Task 6: Game Arena

```python
def play_game(arena):
  #TO DO
  rw,cl=arena.shape
  total=0
  for i in range(rw):
    for j in range(cl):
      if arena[i][j]%50==0 and arena[i][j]>=50:
        if i!=rw-1 and arena[i+1][j]:
           total+=1
        if   i!=0 and arena[i-1][j]==2:
            total+=1
        if j!=cl-1 and arena[i][j+1]==2:
            total+=1
        if   j!=0 and arena[i][j-1]==2:
           total+=1



  sum=(total)*2
  print('Points Gained:',sum,end=' ')
  if sum<10:
    print('.Your team is out.')
  else:
    print('.Your team has survived the game.')




arena=np.array([[0,2,2,0],
                [50,1,2,0],
                [2,2,2,0],
                [1,100,2,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 6. Your team is out.

print(".....................")
arena=np.array([[0,2,2,0,2],
                [1,50,2,1,100],
                [2,2,2,0,2],
                [0,200,2,0,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 14. Your team has survived the game.
```

Bonus Task: Primary vs Secondary Diagonal

```python
def check_diagonal(matrix1, matrix2):
  #TO DO
  row1,col1=matrix1.shape
  row2,col2=matrix2.shape
  sum=0
  sum1=0
  for i in range(row1):
    for j in range(col1):
      if i+j==col1-1:
        sum+=matrix1[i][j]
  for i in range(row2):
    for j in range(col2):
      if i==j:
        sum1+=matrix2[i][j]
  if sum==sum1:
    print('YES')
  else:
    print('NO')
array1 = np.array([[0, 4, 1], [7, 2, 5], [3, 6, 0]])
array2 = np.array([[3, 6, 0], [5, 2, 7], [0, 4, 1]])

check_diagonal(array1, array2) #This should print YES
print(".............")
array1 = np.array([[0, 9, 9, 1], [9, 0, 2, 9], [9, 3, 0, 9], [4, 9, 9, 0]])
array2 = np.array([[4, 9, 9, 0], [9, 0, 3, 9], [9, 0, 2, 9], [0, 9, 5, 1]])

check_diagonal (array1, array2) #This should print NO
```
