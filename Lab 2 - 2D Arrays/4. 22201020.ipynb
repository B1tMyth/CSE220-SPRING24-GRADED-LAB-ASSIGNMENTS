Task 3: Matrix Manipulation

```python
def reverse_Matrix(matrix):
  row,col = matrix.shape
  new_mat = np.zeros((row,col),dtype=int)
  for i in range(row):
    c=0
    for j in range(col-1,-1,-1):
      new_mat[i][c]= matrix[i][j]
      c+=1

  for i in range(col):
    for j in range(row//2):
      new_mat[j][i],new_mat[row-1-j][i]=new_mat[row-1-j][i],new_mat[j][i]
  return new_mat


matrix = np.array([
[14,  8,  0,  4],
[9,  8,  13,  13],
[9,  3,  1,  4],
[2,  10,  13,  6]
])
print_matrix(matrix)
print('################')
print()
reversed_matrix = reverse_Matrix(matrix)
print_matrix(reversed_matrix)

#This should print
#|  6  |  13 |  10 |  2  |
#-------------------------
#|  4  |  1  |  3  |  9  |
#-------------------------
#|  13  |  13  |  8 |  9 |
#-------------------------
#|  4 |  0  |  8  |  14  |
#-------------------------
```

Task 5: Matrix Compression

```python
def compress_matrix(mat):
  row,col = mat.shape
  x,y=int(row/2),int(col/2)
  new_arr = np.zeros((x,y),dtype=int)
  for i in range(0,row,2):
    for j in range(0,col,2):
      sum = mat[i][j]+mat[i][j+1]+mat[i+1][j]+mat[i+1][j+1]
      new_arr[i//2][j//2]=sum
      sum=0
  return new_arr

matrix=np.array([[1,2,3,4],
                 [5,6,7,8],
                 [1,3,5,2],
                 [-2,0,6,-3]])
print_matrix(matrix)
print('################')
print()
returned_array=compress_matrix(matrix)
print_matrix(returned_array)
#This should print
#|  14  |  22 |
#--------------
#|  2  |  10  |
#--------------
```
