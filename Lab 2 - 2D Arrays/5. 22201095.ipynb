```python
def reverse_Matrix(matrix):
   row,col=matrix.shape
   for i in range(row):
     for j in range(col//2):
       matrix[i][j],matrix[row-1-j][i]=matrix[row-1-j][i],matrix[i][j]
   for i in range(col):
     for i in range(row//2):
        matrix[i][j],matrix[row-1-j][i]=matrix[row-1-j][i],matrix[i][j]
   return matrix




matrix = np.array([
[14,  8,  0,  4],
[9,  8,  13,  13],
[9,  3,  1,  4],
[2,  10,  13,  6]
])
print_matrix(matrix)
print('################')
print()
reversed_matrix = reverse_Matrix(matrix)
print_matrix(reversed_matrix)

#This should print
#|  6  |  13 |  10 |  2  |
#-------------------------
#|  4  |  1  |  3  |  9  |
#-------------------------
#|  13  |  13  |  8 |  9 |
#-------------------------
#|  4 |  0  |  8  |  14  |
#-------------------------
```

Task 5: Matrix Compression

```python
def compress_matrix(mat):
    #TO DO
    row,col=mat.shape
    line=np.zeros((row//2,col//2),dtype=int)
    for i in range(0,row,2):
      for j in range(0,col,2):
        line[i//2][j//2]=mat[i][j]+mat[i+1][j]+mat[i][j+1]+mat[i+1][j+1]
    return line


matrix=np.array([[1,2,3,4],
                 [5,6,7,8],
                 [1,3,5,2],
                 [-2,0,6,-3]
                 ])
print_matrix(matrix)
print('################')
print()
returned_array=compress_matrix(matrix)
print_matrix(returned_array)
#This should print
#|  14  |  22 |
#--------------
#|  2  |  10  |
#--------------
```

Task 6: Game Arena

```python

def play_game(arena):
  #TO DO
  row,col=arena.shape
  sum=0
  for i in range(row):
    for j in range(col):
      if arena[i][j]%50==0:
        if arena[i][j]!=0:

          if 0<=i-1<row and arena[i-1][j]==2 :
            sum+=2
          if 0<=i+1<row and arena[i+1][j]==2 :
            sum+=2
          if 0<=j-1<col and arena[i][j-1]==2 :
            sum+=2
          if 0<=j+1<col and arena[i][j+1]==2:
            sum+=2

  if sum>int(10):
    print("Points Gained: ",sum,". Your team has survived the game.")
  else:
    print("Points Gained: ",sum,". Your team is out.")

arena=np.array([[0,2,2,0],
                [50,1,2,0],
                [2,2,2,0],
                [1,100,2,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 6. Your team is out.

print(".....................")
arena=np.array([[0,2,2,0,2],
                [1,50,2,1,100],
                [2,2,2,0,2],
                [0,200,2,0,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 14. Your team has survived the game.
```

Bonus Task: Primary vs Secondary Diagonal

```python

import numpy as np
def check_Diagonal(matrix1, matrix2):
  arr1=matrix1
  arr2=matrix2
  r1=len(arr1)
  r2=len(arr2)
  n_arr1=np.zeros(r1, dtype=int)
  n_arr2=np.zeros(r2, dtype=int)

  for i in range (r1):
      n_arr1[i]=arr1[i][r1-1-i]

  ind=r2-1
  for j in range (r2):
      n_arr2[ind]=arr2[j][j]
      ind-=1
  flag='Yes'
  for i in range(r1):
    if n_arr1[i]==n_arr2[i]:
      flag='Yes'
    else :
      flag='No'
      break
  print(flag)

array1 = np.array([[0, 4, 1], [7, 2, 5], [3, 6, 0]])
array2 = np.array([[3, 6, 0], [5, 2, 7], [0, 4, 1]])

check_Diagonal(array1, array2) #This should print YES
print(".............")
array1 = np.array([[0, 9, 9, 1], [9, 0, 2, 9], [9, 3, 0, 9], [4, 9, 9, 0]])
array2 = np.array([[4, 9, 9, 0], [9, 0, 3, 9], [9, 0, 2, 9], [0, 9, 5, 1]])

check_Diagonal (array1, array2) #This should print NO
```
