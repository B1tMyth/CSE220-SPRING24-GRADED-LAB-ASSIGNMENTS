```python
#Assignment Part 1

class Patient:
  #write a constructor
  def __init__(self, id, name, age, blood, prev, next):
      self.id = id
      self.name = name
      self.age = age
      self.blood = blood
      self.prev = prev
      self.next = next
```

```python
class WRM:
  #write a constructor
  def __init__(self):
     self.dh = Patient(None, None, None, None)
     self.dh.next = self.dh
     self.dh.prev = self.dh
     self.temp = self.dh

  def registerPatient(self,id, name, age, bloodgroup):
    #To Do
    p = Patient(id, name, age, bloodgroup)
    self.temp.next = p
    p.prev = self.temp
    p.next = self.dh
    self.dh.prev = p


  def servePatient(self):
    #To Do
    if self.dh.next != self.dh:
      serve = self.dh.next
      nxt = serve.next
      print(f'{serve.elem[1]} is served')
      self.dh.next = nxt
      nxt.prev = self.dh
    else:
      print('No patient to be served')


  def showAllPatient(self):
    #To Do
    a = self.dh.next
    if self.dh.next != self.dh:
      while a != self.dh:
        print(a.elem[1], end=' ')
        a = a.next

  def canDoctorGoHome(self):
    #To Do
    if self.dh.next == self.dh:
      return 'Yes'
    else:
      return 'No'

  def cancelAll(self):
    #To Do
    self.dh.next = self.dh
    self.dh.prev = self.dh
    print('All appoinments cancelled')


  def ReverseTheLine(self):
    #To Do
    head = self.dh
    v = head.next
    head.next = head.prev
    head.prev = v
    head = head.next
    while head != self.dh:
      v = head.next
      head.next = head.prev
      head.prev = v
      head = head.prev
```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

Task 1: Diamond Count

```python
def diamond_count(stack,string):
  #TO DO
  diamond = 0
  brackets = 0
  for char in string:
      if char == '<':
          stack.push(char)
          brackets += 1
      elif char == '>' and brackets > 0:
          stack.pop()
          diamond += 1
          brackets -= 1
  return diamond




print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

Task 2: Tower of Blocks

```python
def remove_block(stack, n):
  temp = Stack()

  for _ in range(n - 1):
      if stack.peek() is not None:
          temp.push(stack.pop())

  if stack.peek() is not None:
      stack.pop()

  while temp.peek() is not None:
      stack.push(temp.pop())



print('Test 01')
st = Stack()
st.push(4)
st.push(19)
st.push(23)
st.push(17)
st.push(5)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,2)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()

print('Test 02')
st = Stack()
st.push(73)
st.push(85)
st.push(15)
st.push(41)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,3)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()
```

Task 3: Stack Reverse

```python
def conditional_reverse(stack):
  #To Do
  outpt = Stack()
  prev = None

  while stack.peek() is not None:
      current = stack.pop()
      if current != prev:
          outpt.push(current)
      prev = current

  return outpt



print('Test 01')
st=Stack()
st.push(10)
st.push(10)
st.push(20)
st.push(20)
st.push(30)
st.push(10)
st.push(50)
print('Stack:')
print_stack(st)
print('------')
reversed_stack=conditional_reverse(st)
print('Conditional Reversed Stack:')
print_stack(reversed_stack) # This stack contains 50, 10, 30, 20, 10 in this order whereas top element should be 10
print('------')
```
