```python
#Assignment Part 1

class Patient:
  def __init__(self,id,name,age,blood_group):
    self.id=id
    self.name=name
    self.age=age
    self.blood_group=blood_group
class DoublyNode:
  def __init__(self,elem,next=None,prev=None):
    self.elem=elem
    self.next=next
    self.prev=prev
```

```python
class WRM:
  def __init__(self):
    self.head =DoublyNode(None)
    self.head.next=self.head
    self.head.prev=self.head
  def registerPatient(self,id,name,age,bloodgroup):
    new_patient= Patient(id,name,age,bloodgroup)
    new_node= DoublyNode(new_patient)
    new_node.next=self.head
    new_node.prev=self.head.prev
    self.head.prev.next=new_node
  def servePatient(self):
    if self.head.next == self.head:
      return "No patients in the waiting room"
    else:
      served_patient = self.head.next.elem
      self.head.next=self.head.next.next
      self.head.next.prev=self.head
      return f"Serving Patient : {served_patient.name}"
  def cancelAll(self):
    self.head.next=self.head
    self.head.prev=self.head
    return "All appoinments canceled."
  def canDoctorGoHome(self):
    if self.head.next ==self.head:
      return True
    else:
      return False
  def showAllPatients(self):
    current =self.head.next
    while current!=self.head:
      print(current.elem.id,current.elem.name)
      current =current.next
  def reverseTheLine(self):
    if self.head.next ==self.head:
      return "No patients in the waiting room"
    else:
      current = self.head.next
      while current!= self.head.next:
        current.next,current.prev=current.prev,current.next
        current= current.prev
      self.head.next,self.head.prev=self.head.prev,self.head.next
      return "Line reversed succesfully."
```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

Task 1: Diamond Count

```python
def diamond_count(stack,string):
  count=0
  for i in string:
    if i=="<":
      stack.push(i)
    elif i == ">" and stack.peek()=="<":
      stack.pop()
      count+=1
  return count

print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

Task 2: Tower of Blocks

```python
def remove_block(stack, n):
  temp=Stack()
  for i in range(n-1):
    temp.push(stack.pop())
  stack.pop()
  while not temp.isEmpty():
    stack.push(temp.pop())
  return Stack()

print('Test 01')
st = Stack()
st.push(4)
st.push(19)
st.push(23)
st.push(17)
st.push(5)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,2)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()

print('Test 02')
st = Stack()
st.push(73)
st.push(85)
st.push(15)
st.push(41)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,3)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()
```

Task 3: Stack Reverse

```python
def conditional_reverse(stack):
    temp=Stack()
    while stack.peek()!=None:
      x=stack.pop()
      if temp.peek()!=x:
        temp.push(x)
    return temp

print('Test 01')
st=Stack()
st.push(10)
st.push(10)
st.push(20)
st.push(20)
st.push(30)
st.push(10)
st.push(50)
print('Stack:')
print_stack(st)
print('------')
reversed_stack=conditional_reverse(st)
print('Conditional Reversed Stack:')
print_stack(reversed_stack) # This stack contains 50, 10, 30, 20, 10 in this order whereas top element should be 10
print('------')
```
