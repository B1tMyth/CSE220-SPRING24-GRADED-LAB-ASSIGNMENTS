```python
#Assignment Part 1

class Patient:
  #write a constructor
  def __init__(self,id,nam,age,bg):
    self.id=id
    self.name=nam
    self.age=age
    self.bloodgroup=bg
    self.next=None
    self.prev=None
```

```python
class WRM:
  #write a constructor
  def __init__(self):
    self.head=Patient(None,None,None,None)
    self.temp=self.head
    self.temp.next=self.head
    self.temp.prev=self.head

  def registerPatient(self,id, name, age, bloodgroup):
    n=Patient(id,name,age,bloodgroup)
    self.temp.next=n
    n.prev=self.temp
    n.next=self.head
    self.head.prev=n
    self.temp=n

  def servePatient(self):
      if(self.temp==self.head):
        return 'No patient in the waiting room'
      else:
        temp=self.head.next
        save=temp.next
        print(f'Id : {temp.id} , Name :{temp.name}')
        self.head.next=save
        save.prev=self.head
        temp.prev=None
        temp.next=None

  def showAllPatient(self):
    temp=self.head.next
    if(temp==self.head):
      print('No patient in the waiting room')
    else:
      print("Waiting patients ID's:")
      while(temp!=self.head):
        print(f'Id : {temp.id} ')
        temp=temp.next

  def canDoctorGoHome(self):
    temp=self.head.next
    if(temp.id==None):
      print(True)
    else:
      print(False)

  def cancelAll(self):
    self.head.next=self.head
    self.head.prev=self.head
    self.temp=self.head
    print('All appointments canceled. Doctor can go to lunch.')

  def ReverseTheLine(self):
    temp=self.head.prev
    print('reverses the patient line:')
    while(temp!=self.head):
      print(f"ID : {temp.id} , Name : {temp.name}")
      temp=temp.prev
```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

Task 1: Diamond Count

```python
def diamond_count(stack,string):
  count=0
  for i in string:
    if(i=='<'):
      stack.push(i)

    if(i=='>'):
      if(stack.peek()=='<'):
        count+=1
        stack.pop()
  return count


print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

```python
def conditional_reverse(stack):
  #To Do
  st=Stack()

  while stack.isEmpty() == False:
      value = stack.pop()
      st.push(value)

  pre_value = None
  while st.isEmpty() == False:
      value = st.pop()
      if value != pre_value:
        stack.push(value)
        pre_value = value

  return stack


print('Test 01')
st=Stack()
st.push(10)
st.push(10)
st.push(20)
st.push(20)
st.push(30)
st.push(10)
st.push(50)
print('Stack:')
print_stack(st)
print('------')
reversed_stack=conditional_reverse(st)
print('Conditional Reversed Stack:')
print_stack(reversed_stack) # This stack contains 50, 10, 30, 20, 10 in this order whereas top element should be 10
print('------')
```
