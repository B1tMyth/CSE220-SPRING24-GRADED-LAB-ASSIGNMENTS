```python
#Assignment Part 1

class Patient:
  def __init__(self,id,name,age,bloodgroup,prev,next):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup
    self.prev= prev
    self.next= next
```

```python
class WRM:
  def __init__(self):
    self.head = Patient(None,None,None,None,None,None)
    self.head.prev = self.head
    self.head.next = self.head
  def registerPatient(self,id, name, age, bloodgroup):
    new_patient = Patient(id, name, age, bloodgroup,self.head.prev,self.head)
    self.head.prev.next = new_patient
    self.head.prev = new_patient

  def servePatient(self):
    x = self.head.next
    self.head.next = self.head.next.next
    self.head.next.prev = self.head

  def showAllPatient(self):
    temp = self.head
    while True:
      if temp.id!=None:
        print(f"Name: {temp.name}\nID:{temp.id}\nAge:{temp.age}\nBloodgroup{temp.bloodgroup}")
        print()
      if temp.next==self.head:
        break
      temp = temp.next
  def canDoctorGoHome(self):
    temp = self.head
    if temp.next!=self.head:
      return False
    return True
  def cancelAll(self):
    x = self.head.next
    self.head.next = self.head
    self.head.prev = self.head
    x = None
  def ReverseTheLine(self):
      temp = self.head
      while True:
        temp.next,temp.prev = temp.prev,temp.next
        if temp.next == self.head:
          break
        temp = temp.next

hospital = WRM()
hospital.registerPatient(1,"Nafi",17,"A+")
hospital.registerPatient(2,"kafi",15,"A-")
hospital.registerPatient(3,"Pafi",18,"O+")
hospital.registerPatient(4,"Dafi",19,"O+")
hospital.registerPatient(5,"Bafi",10,"O-")
hospital.registerPatient(6,"Gafi",13,"AB+")
hospital.registerPatient(7,"Vafi",15,"AB-")
hospital.registerPatient(8,"Bufi",27,"A+")
hospital.registerPatient(9,"Tafi",11,"A-")
hospital.registerPatient(10,"Gefi",20,"A+")
hospital.showAllPatient()
print(hospital.canDoctorGoHome())
print()
hospital.servePatient()
hospital.ReverseTheLine()
hospital.showAllPatient()
print()
hospital.cancelAll()
hospital.showAllPatient()
print(hospital.canDoctorGoHome())
```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

Task 1: Diamond Count

```python
def diamond_count(stack,string):
  diamond_count = 0
  for i in string:
    if i == "<":
      stack.push(i)
    elif i == ">":
      x = stack.peek()
      if x=="<":
        diamond_count+=1
        stack.pop()
  return diamond_count

print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

Task 2: Tower of Blocks

```python
def remove_block(stack, n):
  new_stack = Stack()
  for i in range(1,n+1):
    if i==n:
      elem = stack.pop()
      elem = None
    else:
      elem = stack.pop()
      new_stack.push(elem)
  while new_stack.peek()!=None:
    elem = new_stack.pop()
    stack.push(elem)
  return stack


print('Test 01')
st = Stack()
st.push(4)
st.push(19)
st.push(23)
st.push(17)
st.push(5)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,2)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()

print('Test 02')
st = Stack()
st.push(73)
st.push(85)
st.push(15)
st.push(41)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,3)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()
```

Task 3: Stack Reverse

```python
def conditional_reverse(stack):
  reversed_stack = Stack()
  consecutive_number = None
  while stack.peek()!=None:
    current_elem = stack.pop()
    if current_elem != consecutive_number:
      reversed_stack.push(current_elem)
      consecutive_number = current_elem
  return reversed_stack



print('Test 01')
st=Stack()
st.push(10)
st.push(10)
st.push(20)
st.push(20)
st.push(30)
st.push(10)
st.push(50)
print('Stack:')
print_stack(st)
print('------')
reversed_stack=conditional_reverse(st)
print('Conditional Reversed Stack:')
print_stack(reversed_stack) # This stack contains 50, 10, 30, 20, 10 in this order whereas top element should be 10
print('------')
```
