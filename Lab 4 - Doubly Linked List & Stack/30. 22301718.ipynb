```python
#Assignment Part 1
class Patient:#node
  #write a constructor
  def __init__(self,name,id,age,bg,next = None,prev = None):
    self.name = name
    self.age = age
    self.id = id
    self.bg = bg
    self.next = next
    self.prev = prev

```

```python
class WRM:
  #write a constructor
  def __init__(self):
    self.dummy_head = Patient(None,None,None,None)
    self.tail = self.dummy_head
    self.dummy_head.next = self.dummy_head



  def RegisterPatient(self):
    #To Do
    id = int(input("Enter ID: "))
    name = input("Enter name: ")
    age = int(input("Enter age: "))
    bloodgroup = input("Enter blood group: ")
    patient_object = Patient(id,name,age,bloodgroup)
    self.tail.next = patient_object
    patient_object.prev = self.tail
    self.tail = patient_object
    self.tail.next = self.dummy_head
    print("Patient has been registered")
  def ServePatient(self):
    #To Do
    curr = self.dummy_head

    if curr.next is self.dummy_head:
      print("No patients left to serve")

    else:
      temp = curr.next

      if temp.next is self.dummy_head:
        print(f"{temp.name} has been served")
        self.dummy_head.next = None
        temp.next = None
        temp.prev = None
        curr.next = self.dummy_head

      else:
        print(f"{temp.name} has been served")
        another_temp = temp.next
        curr.next = another_temp
        another_temp.prev = curr
        temp.next = None
        temp.prev = None

  def ShowAllPatient(self):
    #To Do
    curr = self.dummy_head

    if curr.next is not self.dummy_head:
      temp = curr.next

      while temp is not self.dummy_head:
        print(temp.name,end = ' ')
        temp = temp.next

    else:
      print("No patients registered yet")


  def CanDoctorGoHome(self):
    #To Do
    curr = self.dummy_head

    if curr.next is self.dummy_head:
      print(True)

    else:
      print(False)


  def CancelAll(self):
    #To Do
    curr = self.dummy_head
    curr.next = self.dummy_head
    curr.prev = None
    print("All appointments have been cancelled")

  def ReverseTheLine(self):
    curr = self.dummy_head
    prev = None

    while (curr is not None):
      next_node = curr.next
      curr.next = prev
      curr.prev = next_node
      prev = curr
      curr = next_node

    self.dummy_head = prev
    self.tail = prev

    print("Patient list reversed")
running = True
object_1 = WRM()

while running:
  print("**Choose an option**")
  print("1. RegisterPatient")
  print("2. ServePatient")
  print("3. CancelAll")
  print("4. ShowAllPatients")
  print("5. CanDoctorGoHome")

  print("6. ReverseTheLine")
  print("7. Exit")
  print()
  user = int(input("Enter your choice: "))

  if user == 1:
      object_1.RegisterPatient()

  elif user == 2:
      object_1.ServePatient()

  elif user == 3:
      object_1.CancelAll()

  elif user == 4:
      object_1.ShowAllPatient()

  elif user == 5:
      object_1.CanDoctorGoHome()
  elif user==6:
    object_1.ReverseTheLine()

  elif user == 7:
      running = False

  else:
      print("Invalid number")
  print()

print("**Welcome to Waiting Room Management System**")

```

**Assignment Part 2: Stack**

Linked List based Stack is implemented in the following cell.

```python
class Node:
  def __init__(self,elem=None,next=None):
    self.elem = elem
    self.next = next

class Stack:
  def __init__(self):
    self.__top = None

  def push(self,elem):
    nn = Node(elem,self.__top)
    self.__top = nn

  def pop(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    e = self.__top
    self.__top = self.__top.next
    e.next = None
    return e.elem

  def peek(self):
    if self.__top == None:
      #print('Stack Underflow')
      return None
    return self.__top.elem

  def isEmpty(self):
    return self.__top == None
```

Task 1: Diamond Count

```python
def diamond_count(stack,string):
  #TO DO
  count=0
  for i in string:
    if i=="<":
      stack.push(i)
    elif i==">" and not stack.isEmpty():
      stack.pop()
      count+=1
  return count


print('Test 01')
stack = Stack()
string = '<..><.<..>> '
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')


print('Test 02')
stack = Stack()
string = '<<<..<......<<<<....>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 1
unittest.output_test(returned_value, 1)
print('-----------------------------------------')


print('Test 03')
stack = Stack()
string = '>>><...<<..>>...>...>>>'
returned_value = diamond_count(stack,string)
print(f'Number of Diamonds: {returned_value}') #This should print 3
unittest.output_test(returned_value, 3)
print('-----------------------------------------')
```

Task 2: Tower of Blocks

```python
def remove_block(stack, n):
  temp= Stack()

  for i in range (1,n):
    remove=stack.pop()
    temp.push(remove)
  stack.pop()
  while not temp.isEmpty():
    stack.push(temp.pop())


print('Test 01')
st = Stack()
st.push(4)
st.push(19)
st.push(23)
st.push(17)
st.push(5)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,2)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()

print('Test 02')
st = Stack()
st.push(73)
st.push(85)
st.push(15)
st.push(41)
print('Stack:')
print_stack(st)
print('------')
remove_block(st,3)
print('After Removal')
print_stack(st)
print('------')

print()
print('======================================')
print()
```

Task 3: Stack Reverse

```python
def conditional_reverse(stack):
  #To Do
  temp=Stack()
  while not stack.isEmpty():
    remove=stack.pop()
    temp.push(remove)
    if temp.peek()==stack.peek():
      temp.pop()
    else:
      continue
  return temp

print('Test 01')
st=Stack()
st.push(10)
st.push(10)
st.push(20)
st.push(20)
st.push(30)
st.push(10)
st.push(50)
print('Stack:')
print_stack(st)
print('------')
reversed_stack=conditional_reverse(st)
print('Conditional Reversed Stack:')
print_stack(reversed_stack) # This stack contains 50, 10, 30, 20, 10 in this order whereas top element should be 10
print('------')
```
